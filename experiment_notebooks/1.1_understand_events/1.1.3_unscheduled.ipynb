```python
# CATEGORY
print(df_unscheduled_events['CATEGORY'].unique())
print(len(df_unscheduled_events['CATEGORY'].unique()))
print(df_unscheduled_events['CATEGORY'].value_counts())
```

```python
# convert to list for labels column from string
import ast
df_unscheduled_events['LABELS'] = df_unscheduled_events['LABELS'].apply(ast.literal_eval)

labels_count = {}
for label in df_unscheduled_events['LABELS']:
    for label_child in label:
        if label_child in labels_count:
            labels_count[label_child] += 1
        else:
            labels_count[label_child] = 1

display(labels_count)
print(len(labels_count))
```
