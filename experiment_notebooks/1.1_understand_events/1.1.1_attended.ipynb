```python
# Time distriubtion
import matplotlib.pyplot as plt

# Convert EVENT_START and EVENT_END to datetime
df_attended_events['EVENT_START'] = pd.to_datetime(df_attended_events['EVENT_START'])
df_attended_events['EVENT_END'] = pd.to_datetime(df_attended_events['EVENT_END'])

# Calculate duration in minutes
df_attended_events['DURATION'] = (df_attended_events['EVENT_END'] - df_attended_events['EVENT_START']).dt.total_seconds() / 60

# Group by CATEGORY and calculate mean duration
grouped = df_attended_events.groupby('CATEGORY')['DURATION'].mean().reset_index()

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(grouped['CATEGORY'], grouped['DURATION'], color='skyblue')
plt.title('Average Event Duration by Category (in minutes)')
plt.xlabel('Category')
plt.ylabel('Average Duration (minutes)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Convert the EVENT_START column to datetime
df_attended_events['EVENT_START'] = pd.to_datetime(df_attended_events['EVENT_START'])

# Create a new DataFrame to count events per category and start date
df_time_distribution = df_attended_events.groupby(['CATEGORY', df_attended_events['EVENT_START'].dt.date]).size().reset_index(name='EVENT_COUNT')

# Plotting
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")

# Create a scatterplot for each category with decreased marker size
sns.scatterplot(data=df_time_distribution, x='EVENT_START', y='EVENT_COUNT', hue='CATEGORY', marker='o', s=20)  # 's' controls marker size

plt.title('Time Distribution of Events by Category')
plt.xlabel('Event Start Date')
plt.ylabel('Number of Events')
plt.xticks(rotation=45)
plt.legend(title='Category')
plt.tight_layout()
plt.show()
```

```python
# Convert the EVENT_START and EVENT_END columns to datetime
df_attended_events['EVENT_START'] = pd.to_datetime(df_attended_events['EVENT_START'])
df_attended_events['EVENT_END'] = pd.to_datetime(df_attended_events['EVENT_END'])

# Create a DataFrame to hold each hour for each event
hourly_data = []

for _, row in df_attended_events.iterrows():
    # Create a range of hours between EVENT_START and EVENT_END
    hours_range = pd.date_range(start=row['EVENT_START'], end=row['EVENT_END'], freq='H')
    for hour in hours_range:
        hourly_data.append({'HOUR': hour, 'CATEGORY': row['CATEGORY']})

# Create a new DataFrame
df_hourly_distribution = pd.DataFrame(hourly_data)

# Count occurrences of events for each hour and category
df_hourly_count = df_hourly_distribution.groupby(['CATEGORY', df_hourly_distribution['HOUR'].dt.hour]).size().reset_index(name='EVENT_COUNT')

# Plotting
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")

# Create a barplot for each category
sns.barplot(data=df_hourly_count, x='HOUR', y='EVENT_COUNT', hue='CATEGORY', ci=None)

plt.title('Hourly Distribution of Events by Category (Including Event Duration)')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Events')
plt.xticks(range(0, 24))  # Set x-ticks for each hour
plt.legend(title='Category')
plt.tight_layout()
plt.show()
```
