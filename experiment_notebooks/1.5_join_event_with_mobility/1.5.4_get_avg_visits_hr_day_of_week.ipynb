```python
import pandas as pd 
import os
# Set the maximum number of rows and columns to display
pd.set_option('display.max_rows', 200)  # Adjust this number as needed
pd.set_option('display.max_columns', 50)  # Adjust this number as needed
```

```python
def get_weekly_day_hour(df):
    # Convert EVENT_START_BOSTON column to datetime
    df['EVENT_START_BOSTON'] = pd.to_datetime(df['EVENT_START_BOSTON'])
    
    df['START_HOUR'] = df['EVENT_START_BOSTON'].dt.hour

    # Add a new column for the day of the week
    df['DAY_OF_WEEK'] = df['EVENT_START_BOSTON'].dt.day_name()
    return df

def get_event_duration(df):
    df['EVENT_START_BOSTON'] = pd.to_datetime(df['EVENT_START_BOSTON'])
    df['EVENT_END_BOSTON'] = pd.to_datetime(df['EVENT_END_BOSTON'])

    df['EVENT_DURATION'] = df['EVENT_END_BOSTON'] - df['EVENT_START_BOSTON']

    return df
```

```python
import os
# List all files in the folder
folder_path = '../../data_CityEvent/processed/1.5.3_events_join_w_mobility/'
file_names = os.listdir(folder_path)

# file_names.remove('unscheduled_intermediate') 
# file_names.remove('archive_before_dropping_missing_geometry') 
for file_name in file_names:
    print(file_name)
```

```python
import os
import pandas as pd
for file_name in file_names:
    current_df = pd.read_csv(os.path.join(folder_path, file_name))

    current_df = get_event_duration(current_df)
    print(f'After event duration transformation: {current_df.shape}')
    current_df = get_weekly_day_hour(current_df)
    print(f'After weekly-hour transformation: {current_df.shape}')

    # Group by DAY_OF_WEEK, START_HOUR, and BGFIPS
    result = current_df.groupby(['DAY_OF_WEEK', 'START_HOUR', 'BGFIPS']).agg(
        TOTAL_VISITS=('Visits_Mobility', 'sum'),
        AVG_VISITS=('Visits_Mobility', 'mean'),
        EVENTS_COUNT=('EVENT_ID', 'count')  # Count the number of events
    ).reset_index()
    
    # print shape
    print(f'Result shape: {result.shape}')

    display(result.head())   

    # # save result
    result.to_csv(f'../../data_CityEvent/processed/1.5.4_avg_visits_across_day_of_week_at_cbgid_hour_of_day/avg_visits_{file_name}', index=False)
    
```
